```python
import os
from PIL import Image

folder = 'eigenfaces usfca'
files = os.listdir(folder)

widths = []
heights = []

for file in files:
    if file.endswith('.jpg') or file.endswith('.png'):
        path = os.path.join(folder, file)
        img = Image.open(path)
        w, h = img.size  # width, height
        
        widths.append(w)
        heights.append(h)

print("Minimum width:", min(widths))
print("Minimum height:", min(heights))

# Minimum resolution for images from USFCA staff selfies
```

```python
import os
from PIL import Image

folder = 'eigenfaces usfca'
files = os.listdir(folder)

sizes = set()

for file in files:
    if file.endswith('.jpg') or file.endswith('.png'):
        path = os.path.join(folder, file)
        img = Image.open(path)
        sizes.add(img.size)

print(sizes)
```

```python
import os
import numpy as np
from PIL import Image

folder = 'eigenfaces usfca'
files = os.listdir(folder)

TARGET_SIZE = (1274, 1274)

images = []

for file in files:
    if file.endswith('.jpg') or file.endswith('.png'):
        path = os.path.join(folder, file)
        
        img = Image.open(path)
        img = img.convert('L')            # grayscale
        img = img.resize(TARGET_SIZE)     # enforce min resolution
        
        images.append(np.array(img))

images = np.array(images)

print(images.shape)

# Force minimum basis for all images
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

first_four = [] # storing the four images from selfies
for i in range(4):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    first_four.append(images[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
# Plotting 4 images (Step 1)
```
